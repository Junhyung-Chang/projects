```
import pandas as pd
import matplotlib.pyplot as plt
 
df = pd.read_csv("covid19aug1.csv")
```

```
krdf=df[df['geoId'] == 'KR']
cndf=df[df['geoId'] == 'CN']
jpdf=df[df['geoId'] == 'JP']
usdf=df[df['geoId'] == 'US']
```

```
revcaskr= krdf.loc[::-1, 'cases']
revcasch= cndf.loc[::-1, 'cases']
revcasjp= jpdf.loc[::-1, 'cases']
revcasus= usdf.loc[::-1, 'cases']
```

```
krdf['total']=revcaskr.cumsum()
cndf['total']=revcasch.cumsum()
jpdf['total']=revcasjp.cumsum()
usdf['total']=revcasus.cumsum()
```

```
krscat=plt.scatter(krdf['total'], krdf['cases'], c='blue')
cnscat=plt.scatter(cndf['total'], cndf['cases'], c='green')
jpscat=plt.scatter(jpdf['total'], jpdf['cases'], c='orange')
usscat=plt.scatter(usdf['total'], usdf['cases'], c='purple')

plt.xscale("log")
plt.yscale("log")
plt.xlim((1,10**7))
plt.ylim((0.1,10**6))
plt.title('Covid-19')
ctr=['Korea', 'China', 'Japan', 'US']
plt.legend(ctr ,loc="upper left", title="countries")
```
